```python
# Exercício 1 – Análise de crescimento de empresas

# (a) Divisão por setor e comparação de média e mediana
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados
dados = pd.DataFrame({
    "Meses": [8, 9, 4, 5, 3, 6, 8, 6, 6, 8, 5, 5, 6, 4, 4],
    "Setor": ['C', 'C', 'I', 'I', 'I', 'C', 'C', 'I', 'I', 'C', 'C', 'I', 'C', 'I', 'I'],
    "Tamanho": ['G', 'M', 'G', 'M', 'M', 'P', 'G', 'M', 'P', 'M', 'P', 'P', 'M', 'M', 'G']
})

# Agrupar por setor
setor_stats = dados.groupby("Setor")["Meses"].agg(["mean", "median"])
print("Média e mediana por setor:")
print(setor_stats)


# (b) Desvio padrão e variância por setor
setor_disp = dados.groupby("Setor")["Meses"].agg(["std", "var"])
# O grupo com menor desvio padrão/variância é o mais homogêneo.
print("\nDesvio padrão e variância por setor:")
print(setor_disp)

# (c) Estatísticas por tamanho da empresa (P, M, G)
tamanho_stats = dados.groupby("Tamanho")["Meses"].agg(
    ["mean", "median", "std", "var"])
print("\nEstatísticas por tamanho de empresa:")
print(tamanho_stats)


plt.figure(figsize=(10, 5))

# Boxplot por setor
plt.subplot(1, 2, 1)
sns.boxplot(x="Setor", y="Meses", data=dados)
plt.title("Distribuição de meses com crescimento por Setor")

# Boxplot por tamanho
plt.subplot(1, 2, 2)
sns.boxplot(x="Tamanho", y="Meses", data=dados)
plt.title("Distribuição por Tamanho da Empresa")

plt.tight_layout()
plt.show()
```

```python
# Exercício 2 – Peso por faixa etária (Python + gráfico)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo do gráfico
sns.set(style="whitegrid")

# Dados: 6 faixas etárias × 6 valores cada = 36 valores
idades = [40]*6 + [45]*6 + [50]*6 + [55]*6 + [60]*6 + [65]*6
pesos = [
    55, 50, 68, 65, 62, 58,  # idade 40
    56, 62, 65, 63, 60, 74,  # idade 45
    70, 78, 76, 77, 78, 70,  # idade 50
    72, 80, 70, 76, 74, 83,  # idade 55
    85, 65, 82, 72, 82, 80,  # idade 60
    75, 68, 70, 74, 76, 77   # idade 65
]

# Criando o DataFrame
df2 = pd.DataFrame({"Idade": idades, "Peso": pesos})

# Estatísticas por faixa etária
estatisticas_idade = df2.groupby("Idade")["Peso"].agg(
    ["mean", "median", "std", "var"])
print("📌 Estatísticas por idade:\n")
print(estatisticas_idade)

# Gráfico de linha: peso médio e mediano por idade
plt.figure(figsize=(10, 5))
sns.lineplot(data=estatisticas_idade[[
             'mean', 'median']], marker="o", palette="Set1")
plt.title("Peso médio e mediano por faixa etária")
plt.ylabel("Peso (kg)")
plt.xlabel("Idade (anos)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Exercício 4 – Quartis e histograma (nível de potássio)


import matplotlib.pyplot as plt

# Frequências e classes
classes = ["2.35–2.55", "2.55–2.75", "2.75–2.95",
           "2.95–3.15", "3.15–3.35", "3.35–3.55"]
frequencias = [1, 3, 2, 4, 5, 6]

# Expandir dados para cálculo de quartis
valores_expandido = []
limites = [(2.35, 2.55), (2.55, 2.75), (2.75, 2.95),
           (2.95, 3.15), (3.15, 3.35), (3.35, 3.55)]
for (a, b), freq in zip(limites, frequencias):
    valores_expandido += [(a + b)/2] * freq

# Cálculo dos quartis
q1 = pd.Series(valores_expandido).quantile(0.25)
q2 = pd.Series(valores_expandido).quantile(0.50)
q3 = pd.Series(valores_expandido).quantile(0.75)
print(f"📌 Quartis: Q1={q1:.2f}, Q2={q2:.2f}, Q3={q3:.2f}")

# Histograma
plt.figure(figsize=(8, 4))
plt.bar(classes, frequencias, color='skyblue')
plt.title("Histograma dos níveis de potássio")
plt.xlabel("Nível de potássio")
plt.ylabel("Frequência")
plt.tight_layout()
plt.show()
```

```python
# Exercício 6 – Teste de resistência antes e depois do treinamento

import pandas as pd
import matplotlib.pyplot as plt

# Classes e frequências
classes = ["0–2", "2–4", "4–8", "8–12", "12–16", "16–22"]
antes = [442, 200, 130, 34, 10, 3]
depois = [80, 200, 280, 179, 43, 8]

# Histograma comparativo
x = range(len(classes))
plt.figure(figsize=(10, 5))
plt.bar(x, antes, width=0.4, label='Antes', align='center', color='salmon')
plt.bar([i + 0.4 for i in x], depois, width=0.4,
        label='Depois', align='center', color='seagreen')
plt.xticks([i + 0.2 for i in x], classes)
plt.title("Distribuição dos testes de resistência (antes x depois)")
plt.xlabel("Km percorridos sem parar")
plt.ylabel("Frequência")
plt.legend()
plt.tight_layout()
plt.show()

# Estimativas usando pontos médios
limites = [(0, 2), (2, 4), (4, 8), (8, 12), (12, 16), (16, 22)]


def expand_data(freqs, limites):
    values = []
    for (a, b), f in zip(limites, freqs):
        values += [(a + b)/2] * f
    return pd.Series(values)


antes_series = expand_data(antes, limites)
depois_series = expand_data(depois, limites)

print("📌 Estatísticas antes:")
print(antes_series.describe(percentiles=[.25, .5, .75]))

print("\n📌 Estatísticas depois:")
print(depois_series.describe(percentiles=[.25, .5, .75]))
```

```python
# Exercício 11 – Probabilidade normal (poluente)

from scipy.stats import norm

# Dados: média = 8 ppm, desvio padrão = 1.5 ppm
media = 8
dp = 1.5
limite = 10

# P(X > 10)
prob = norm.sf(limite, loc=media, scale=dp)
print(f" Probabilidade de concentração exceder 10 ppm: {prob:.4%}")
```

```python
# Exercício 12 – Corrente elétrica em fio (distribuição normal)

# Dados: média = 10 mAh, variância = 4 (=> dp = 2)
media = 10
dp = 2

# (a) P(X > 13)
p_a = norm.sf(13, loc=media, scale=dp)
print(f" (a) Probabilidade de exceder 13 mAh: {p_a:.4%}")

# (b) P(9 < X < 11)
p_b = norm.cdf(11, loc=media, scale=dp) - norm.cdf(9, loc=media, scale=dp)
print(f" (b) Probabilidade entre 9 e 11 mAh: {p_b:.4%}")
```

```python
# Exercício 13 – Duração de motores (distribuição normal)
# Dados: média = 150.000 km, dp = 5.000 km

media = 150000
dp = 5000

# (a) P(X < 170000)
p_a = norm.cdf(170000, loc=media, scale=dp)
print(f" (a) Probabilidade de durar menos de 170.000 km: {p_a:.4%}")

# (b) P(140000 < X < 165000)
p_b = norm.cdf(165000, loc=media, scale=dp) - \
    norm.cdf(140000, loc=media, scale=dp)
print(f" (b) Probabilidade entre 140.000 e 165.000 km: {p_b:.4%}")
```
