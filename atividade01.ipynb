```python
'''É esperado que a massa muscular de uma pessoa diminua com a idade. Para estudar
essa relação, uma nutricionista selecionou 18 mulheres, com idade entre 40 e 79 anos,
e observou em cada uma delas a idade (X) e a massa muscular (Y).'''

'''a) Ajuste uma reta de regressão para a relação entre as variáveis Y: massa
muscular (dependente) e X: idade (independente).

(b) Considerando a reta estimada dada no item (a), estime a massa muscular média
de mulheres com 50 anos.
'''

# Importação das bibliotecas necessárias
import numpy as np
import numpy as np
import matplotlib.pyplot as pltr_model import LinearRegression

# Dados fornecidos em listas para o cálculo da regressão
massa_muscular = [82.0, 91.0, 100.0, 68.0, 87.0, 73.0, 78.0, 80.0, 65.0, 84.0,
                  116.0, 76.0, 97.0, 100.0, 105.0, 77.0, 73.0, 78.0]
idade = [71.0, 64.0, 43.0, 67.0, 56.0, 73.0, 68.0, 56.0, 76.0, 65.0,
         45.0, 58.0, 45.0, 53.0, 49.0, 78.0, 73.0, 68.0]

# Convertendo as listas para arrays do numpy
X = np.array(idade).reshape(-1, 1)  # Variável independente (idade)
y = np.array(massa_muscular)        # Variável dependente (massa muscular)

# Ajustando o modelo de regressão linear, fit() resolve automaticamente os coeficientes B0 e B1
modelo = LinearRegression()
modelo.fit(X, y)

# Coeficientes da reta de regressão
coef_angular = modelo.coef_[0]  # Inclinação da reta
intercepto = modelo.intercept_  # Intercepto (onde a reta cruza o eixo y)

# Exibindo a equação da reta de regressão
print(
    f"Equação da reta: massa_muscular = {intercepto:.2f} + ({coef_angular:.2f}) * idade")

# Estimando a massa muscular para uma mulher de 50 anos
# A entrada precisa ser um array 2D para a previsão
idade_50 = np.array([[50]])
massa_estimada_50 = modelo.predict(idade_50)[0]
print(
    f"Massa muscular estimada para uma mulher de 50 anos: {massa_estimada_50:.2f} kg")

# Plotando os dados e a reta de regressão
plt.scatter(X, y, color='blue', label='Dados observados')  # Dados originais
plt.plot(X, modelo.predict(X), color='red',
         label='Reta de regressão')  # Reta de regressão
plt.scatter(50, massa_estimada_50, color='green',
            label='Estimativa para 50 anos')  # Estimativa para 50 anos
plt.xlabel("Idade (anos)")  # Rótulo do eixo X
plt.ylabel("Massa Muscular (kg)")  # Rótulo do eixo Y
plt.title("Regressão Linear: Idade vs Massa Muscular")  # Título do gráfico
plt.legend()  # Exibe a legenda
plt.grid(True)  # Exibe a grade no gráfico
plt.show()  # Exibe o gráfico
```
